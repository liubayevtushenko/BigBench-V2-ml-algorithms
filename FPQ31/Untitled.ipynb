```python
from pyspark.mllib.util import MLUtils
from pyspark.sql import SQLContext
from pyspark.mllib.fpm import FPGrowth
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.rdd import RDD
```

```python
spark = SparkSession.builder.appName('Project').getOrCreate()
```

```python
dataset=spark.read.csv("store_sales.tbl", inferSchema = True, header = True, sep = '|')
```

```python
q1= "drop view store_sales"
```

```python
spark.sql(q1)
```

```python
dataset.createTempView("store_sales")
```

```python
q="Select collect_set(ss_item_id) FROM store_sales GROUP BY ss_transaction_id"
```

```python
df = spark.sql(q)
```

```python
df1 = df.rdd.map(lambda row: row[0])
```

```python
model = FPGrowth.train(df1, minSupport=0.004, numPartitions=2)
```

```python
result = model.freqItemsets().collect()
```
